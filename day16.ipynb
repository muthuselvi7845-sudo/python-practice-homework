```python
L=[59,1,89,30,2,5]
min=L[0]
max=L[0]

for x in L:
    if x > max:
        max = x
    elif x < min:
        min = x

print("max:", max)
print("min:", min)
```

```python
L = [59, 1, 89, 30, 2, 5]

firstLarge = L[0]
secondLarge = float('-inf')   

for x in L:
    if x > firstLarge:
        secondLarge = firstLarge
        firstLarge = x
    elif x != firstLarge and x > secondLarge:
        secondLarge = x

print("second largest number in the array =", secondLarge)
```
