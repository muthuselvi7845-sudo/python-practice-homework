```python
num1=int(input("Enter a number:"))
exp=(input("Enter a expresion:"))
num2=int(input("Enter a number:"))
if (exp=="+"):
    print(num1+num2)
elif(exp=="-"):
    print(num1-num2)
elif(exp=="*"):
    print(num1*num2)
elif(exp=="/"):
    print(num1/num2)
elif(exp=="**"):
    print(num1**num2)
elif(exp=="%"):
    print(num1*num2)
elif(exp=="//"):
    print(num1//num2)
else:
    print("Not Found")








```
